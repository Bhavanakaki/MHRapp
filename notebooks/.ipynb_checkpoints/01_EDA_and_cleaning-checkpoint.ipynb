```python
import pandas as pd
df = pd.read_csv('../data/survey.csv')
df.head(), df.info()
```

```python
#EDA
import matplotlib.pyplot as plt
df['work_interfere'].value_counts().plot.bar(title='Interference Distribution')
plt.show()
```

```python
#clean+encode target
# Drop missing target
df = df.dropna(subset=['work_interfere'])

# Map to Low/Moderate/High
risk_map = {'Never':'Low','Rarely':'Moderate','Sometimes':'Moderate','Often':'High'}
df['RiskLevel'] = df['work_interfere'].map(risk_map)

# Drop unused cols
df = df.drop(columns=['work_interfere','Timestamp','state','comments','treatment'])
df.to_csv('../data/cleaned_survey.csv', index=False)
print("Saved cleaned_survey.csv")
```

```python
#Visualization: RiskLevel Distribution
import matplotlib.pyplot as plt

df['RiskLevel'].value_counts().plot.bar(color='skyblue')
plt.title("Distribution of Risk Levels")
plt.xlabel("Risk Level")
plt.ylabel("Count")
plt.show()
```

```python
# Visualization: Key Feature Distributions (rectified Age plot)
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1) Age Distribution: 5‑year bins between 20 and 65
bins = list(range(20, 66, 5))  # [20,25,30,...,65]
axes[0].hist(df['Age'].dropna(), bins=bins, edgecolor='black')
axes[0].set_title("Age Distribution")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Number of Respondents")

# 2) Family History
df['family_history'].value_counts().plot.bar(
    ax=axes[1], color='skyblue', edgecolor='black'
)
axes[1].set_title("Family History of Mental Illness")
axes[1].set_xlabel("Family History")
axes[1].set_ylabel("Count")

# 3) Benefits Offered
# ensure consistent order if you like: Yes, No, Don't know
order = ['Yes', 'No', "Don't know"]
df['benefits'].value_counts().reindex(order).plot.bar(
    ax=axes[2], color='lightgreen', edgecolor='black'
)
axes[2].set_title("Employer Mental‑Health Benefits")
axes[2].set_xlabel("Benefits Offered")
axes[2].set_ylabel("Count")

plt.tight_layout()
plt.show()
```
